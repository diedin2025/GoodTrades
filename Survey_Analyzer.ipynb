# 📝 Project 2: Survey Analyzer
**Topics:** Sets, Dictionaries

```python
import random
# Name: Ian Mason
# Task1: Generate Survey Data (20 pts)

# Create a dictionary named survey
survey = {}

# Predefined hobby pool
sample_pool_hobbies = ["Ice Skating", "Reading", "Cooking", "Gaming", "Hiking",
                       "Painting", "Cycling", "Yoga", "Traveling", "Guitar", "Swimming"]

# Name pools to generate full names
first_names = ["Emma", "Liam", "Olivia", "Noah", "Ava", "Elijah", "Sophia", "James",
               "Isabella", "William", "Mia", "Benjamin", "Charlotte", "Lucas", "Amelia",
               "Henry", "Harper", "Alexander", "Evelyn", "Daniel", "Abigail", "Matthew",
               "Emily", "Jack", "Ella", "Samuel", "Grace", "David", "Chloe", "Joseph"]

last_names = ["Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller",
              "Davis", "Rodriguez", "Martinez", "Wilson", "Anderson", "Taylor", "Thomas",
              "Moore", "Jackson", "Martin", "Lee", "Clark", "Lewis", "Walker", "Hall",
              "Allen", "Young", "Wright", "Adams", "Nelson", "Hill", "King", "Scott"]

# Use loop to construct the dictionary with 50 entries
# Each key is a full name, each value is a set of 2 to 4 randomly selected hobbies
while len(survey) < 50:
    first = random.choice(first_names)
    last = random.choice(last_names)
    full_name = f"{first} {last}"
    # Avoid duplicate names
    if full_name not in survey:
        num_hobbies = random.randint(2, 4)
        hobbies = set(random.sample(sample_pool_hobbies, num_hobbies))
        survey[full_name] = hobbies

# Print the resulting dictionary with top 10 entries
print("Survey Data (first 10 entries):")
print("-" * 40)
for i, (name, hobbies) in enumerate(survey.items()):
    if i >= 10:
        break
    print(f"{name}: {', '.join(hobbies)}")
```

## ✅ Task 2: Analyze the Survey Data (30 pts)

```python
# Task 2: Analyze data
# Name:

# Find and print the total number of unique hobbies in survey
all_hobbies = []
for hobbies in survey.values():
    for hobby in hobbies:
        all_hobbies.append(hobby)

unique_hobbies = set(all_hobbies)
print(f"Total number of unique hobbies: {len(unique_hobbies)}")
print(f"Unique hobbies: {unique_hobbies}")

# Find and print the most common hobby and how many people like it
# Count occurrences of each hobby
hobby_counts = {}
for hobby in all_hobbies:
    if hobby in hobby_counts:
        hobby_counts[hobby] += 1
    else:
        hobby_counts[hobby] = 1

# Find the most common hobby
most_common_hobby = max(hobby_counts, key=hobby_counts.get)
most_common_count = hobby_counts[most_common_hobby]

print(f"\nMost common hobby: {most_common_hobby} ({most_common_count} people)")
```

## ✅ Task 3: People Sharing the Same Hobbies (25 pts)

```python
# Name:
# Task 3: People with same hobbies
def group_people_by_hobby(survey):
    hobby_groups = {}
    # Create a reverse mapping: hobby -> set of people
    for name, hobbies in survey.items():
        for hobby in hobbies:
            if hobby not in hobby_groups:
                hobby_groups[hobby] = set()
            hobby_groups[hobby].add(name)
    return hobby_groups

# Print each hobby and the list of names of participants who share it
hobby_groups = group_people_by_hobby(survey)
for hobby, participants in hobby_groups.items():
    print(f"{hobby}: {', '.join(participants)}")
```

## ✅ Task 4: Add a New Participant (25 pts)

```python
# Name:
# Task 4: Add entry and update file
name = input("\nEnter new participant's full name: ")
hobbies_input = input("Enter 2 to 4 hobbies, separated by commas: ")

# Combine and add this entry to the existing dictionary
new_hobbies = set(hobby.strip() for hobby in hobbies_input.split(","))
survey[name] = new_hobbies

# Print the updated dictionary size and verify the new entry appears correctly
print(f"\nNew participant added. Total participants: {len(survey)}")
print(f"{name}: {survey[name]}")
```
